```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
data = np.load('dump/gan_results.npz')
data.files
```

```python
train_source_acc = data['train_source_acc']
train_target_acc = data['train_target_acc']
train_avg_acc = data['train_avg_acc']
val_source_acc = data['val_source_acc']
val_target_acc = data['val_target_acc']
val_avg_acc = data['val_avg_acc']
```

```python
epochs = range(1, len(train_source_acc) + 1)
```

```python
plt.figure(figsize=(20, 10))

# Plotting training accuracies
sns.lineplot(x=epochs, y=train_source_acc, label='Train Source Accuracy', color='blue', linestyle='-')
sns.lineplot(x=epochs, y=val_source_acc, label='Test Source Accuracy', color='blue', linestyle='--')
sns.lineplot(x=epochs, y=train_target_acc, label='Train Target Accuracy', color='red', linestyle='-')
sns.lineplot(x=epochs, y=val_target_acc, label='Test Target Accuracy', color='red', linestyle='--')
sns.lineplot(x=epochs, y=train_avg_acc, label='Train Average Accuracy', color='green', linestyle='-')
sns.lineplot(x=epochs, y=val_avg_acc, label='Test Average Accuracy', color='green', linestyle='--')

# Adding labels and title
plt.title('Training Accuracy Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(title='Accuracy Type')
plt.grid(True)
plt.show()
```
