```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
data = np.load('dump/resnet18_mnist_svhn_results.npz')
data.files
```

```python
data = np.load('dump/resnet18_mnist_usps_results.npz')
res_mnist_usps_val_target_acc = data['val_target_acc']
print("Resnet18 MNIST -> USPS", data['val_source_acc'][-1], data['val_target_acc'][-1])
data = np.load('dump/resnet18_usps_mnist_results.npz')
res_usps_mnist_val_target_acc = data['val_target_acc']
print("Resnet18 USPS -> MNIST", data['val_source_acc'][-1], data['val_target_acc'][-1])


data = np.load('dump/resnet18_multitask_mnist_usps_results.npz')
multi_mnist_usps_val_target_acc = data['val_target_acc']
print("Multi-task MNIST -> USPS", data['val_source_acc'][-1], data['val_target_acc'][-1])
data = np.load('dump/resnet18_multitask_usps_mnist_results.npz')
multi_usps_mnist_val_target_acc = data['val_target_acc']
print("Multi-task USPS -> MNIST", data['val_source_acc'][-1], data['val_target_acc'][-1])


data = np.load('dump/gan_mnist_usps_results.npz')
gan_mnist_usps_val_target_acc = data['val_target_acc']
print("GAN MNIST -> USPS", data['val_source_acc'][-1], data['val_target_acc'][-1])
data = np.load('dump/gan_usps_mnist_results.npz')
gan_usps_mnist_val_target_acc = data['val_target_acc']
print("GAN USPS -> MNIST", data['val_source_acc'][-1], data['val_target_acc'][-1])
```

```python
epochs = range(1, len(res_mnist_usps_val_target_acc) + 1)
```

```python
fig, ax = plt.subplots(ncols=2, figsize=(20, 6))

# Plotting the mnist -> usps results
sns.lineplot(x=epochs, y=res_mnist_usps_val_target_acc, label='Resnet18 Target Accuracy', color='blue', linestyle='-', ax=ax[0])

sns.lineplot(x=epochs, y=multi_mnist_usps_val_target_acc, label='Multi-task Target Accuracy', color='red', linestyle='-', ax=ax[0])

sns.lineplot(x=epochs, y=gan_mnist_usps_val_target_acc, label='GAN Target Accuracy', color='green', linestyle='-', ax=ax[0])

ax[0].legend(title='Accuracy Type')
ax[0].set_title('MNIST -> USPS')
ax[0].set_xlabel('Epochs')
ax[0].set_ylabel('Accuracy')
ax[0].set_ylim([0.4, 1])

# Plotting the usps -> mnist results
sns.lineplot(x=epochs, y=res_usps_mnist_val_target_acc, label='Resnet18 Target Accuracy', color='blue', linestyle='-', ax=ax[1])

sns.lineplot(x=epochs, y=multi_usps_mnist_val_target_acc, label='Multi-task Target Accuracy', color='red', linestyle='-', ax=ax[1])

sns.lineplot(x=epochs, y=gan_usps_mnist_val_target_acc, label='GAN Target Accuracy', color='green', linestyle='-', ax=ax[1])

# Adding labels and title
ax[1].legend(title='Accuracy Type')
ax[1].set_title('USPS -> MNIST')
ax[1].set_xlabel('Epochs')
ax[1].set_ylabel('Accuracy')
ax[1].set_ylim([0.4, 1])
fig.suptitle('Test Accuracy Over Epochs')
plt.grid(True)
plt.show()
```
